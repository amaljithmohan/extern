```python
#removing special characters
import re
exp=r'[A-Za-z0-9]+'
df["title"]=df["title"].apply(lambda x: " ".join(re.findall(exp,x)))
```

```python
import numpy as np

df1=np.array(df['title'])
```

```python
#stopwords and tokenize
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')

stop_words = stopwords.words('english')
  

  
filtered_sentences=[]
for i in df1:
  word_tokens = word_tokenize(i)
  filtered_sentences.append(" ".join([w for w in word_tokens if not w.lower() in stop_words]))
  # break
filtered_sentences
```

```python
#tf 

from collections import Counter
tf_dict={}
for i in range(len(filtered_sentences)):

  count=filtered_sentences[i].split()
  c=Counter(count) 

  for j in Counter(count).keys():
    tf_dict[i,j] =(Counter(count)[j]/len(count))
  


```

```python
tf_dict
```

```python
#idf


from collections import defaultdict
df_dict=defaultdict(list)
def df(wrd):
  wrd=wrd.split()
  for i in range ((len(wrd))):
    
    for j in wrd[i].split(' '):
      
      df_dict[j].append(str(i))
for i in filtered_sentences:
      df(i)

for i in df_dict.keys():
  df_dict[i]=len(df_dict[i])
tf_idf={}
N=len(filtered_sentences)
for i in range (len(filtered_sentences)):
  txt=filtered_sentences[i].split()
  for j in txt:
      tf_v=tf_dict[i,j]
      df=df_dict[j]
      idf = np.log(N/(df+1))
      tf_idf[(i, j)] = tf_v*idf
```

```python
tf_idf
```

```python
#cosine similarity


def create_dataframe(matrix, tokens):
    doc_names = [f'doc_{i+1}' for i, _ in enumerate(matrix)]
    df = pd.DataFrame(data=matrix, index=doc_names, columns=tokens)
    return(df)
simlrty={}
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
for i in range (len(filtered_sentences)):
  doc_2 = "Design Kit Prototyping"
  doc_1=filtered_sentences[i]
  data = [doc_1, doc_2]
  count_vectorizer = CountVectorizer()
  vector_matrix = count_vectorizer.fit_transform(data)
  cosine_similarity_matrix = cosine_similarity(vector_matrix)
  create_dataframe(cosine_similarity_matrix,['doc_1','doc_2'])
  
  simlrty[i]=(cosine_similarity_matrix[0][1])
  
print(dict(sorted(simlrty.items(), key=lambda item: item[1])))
```
