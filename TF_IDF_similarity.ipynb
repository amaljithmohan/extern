```python
#TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_df = 0.3,
                            stop_words="english",
                            lowercase= True,
                            use_idf=True,
                            smooth_idf= True,
                            norm =u'l2')
```

```python
data = df["DETAIL ABOUT MOVIE"]
tfidf_matrix = vectorizer.fit_transform(data)
```

```python
vec = pd.DataFrame(tfidf_matrix[0].T.todense(), index = vectorizer.get_feature_names(), columns=['TF-IDF'])

vec = vec.sort_values('TF-IDF', ascending=False)

vec
```

```python
#Cosine_Similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(tfidf_matrix,tfidf_matrix)
print(dict(sorted(similarity.items(), key=lambda item: item[1])))
```

```python
new_document = "Jake Sully lives with his newfound family formed on the extrasolar moon Pandora. Once a familiar threat returns to finish what was previously started, Jake must work with Neytiri and the army of the Na'vi race to protect their home."
new_document_tfidf = vectorizer.transform([new_document])
new_similarity = cosine_similarity(new_document_tfidf, tfidf_matrix)
  
print(new_similarity)
```
